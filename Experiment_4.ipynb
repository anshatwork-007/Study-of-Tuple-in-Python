## Title: Study of Tuple in Python


```python
# Python Tuples - locks the item, .remove, .append wont be possible
# Access Tuple Items

thistuple = ("Tesla", "Porche", "Audi")
print(thistuple[1])
```

```python
# Slicing in tuple - use '()' instead of '[]'

thistuple = (102, 203, 304, 406, 507, 609,)
print(thistuple[1])
print(thistuple[1:3])
print(thistuple[1:3:1])
print(thistuple[2:5:2])
print(thistuple[1:4:3])
print(thistuple[-1:3:1])
print(thistuple[-1:-3:-1])
```

```python
# To show tuple locks items and shows error if we try to change it
thistuple = ("Tesla", "Porche", "Audi")
thistuple[1] = "Maruti"
print(thistuple)
```

```python
# Append Tuple
thistuple = ("Tesla", "Porche", "Audi")
thistuple.append("Maruti")
print(thistuple)
```

```python
# Adding tuple
t1 = (1, 2, 3)
t2 = (4, 5, 6)
t3 = t1 + t2
print("Tuple 1: ", t1)
print("Tuple 2: ", t2)
print("Tuple 3: ", t3)
```
